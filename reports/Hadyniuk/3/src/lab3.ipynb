```python
from ultralytics import YOLO
```

```python
main_path = "/content/drive/MyDrive/lab3"
dataset_path = main_path + "/dataset"
data_yaml = main_path + "/data.yaml"
model_path = main_path + "/yolov8s.pt"
```

```python
model = YOLO(model_path)
```

```python
model.train(data=data_yaml, epochs=20, imgsz=640, batch=16, device=0, project='runs/train')
```

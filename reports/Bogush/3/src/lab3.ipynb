```python
from ultralytics import YOLO
```

```python
main_path = "/content/drive/MyDrive/Colab Notebooks/lab3-Oivis"
dataset_path = main_path + "/dataset"
data_yaml = main_path + "/data.yaml"
```

```python
model = YOLO('yolov8n.pt')
```

```python
model.train(data = data_yaml, epochs = 20, imgsz = 640, batch = 16, device = 0)
```
